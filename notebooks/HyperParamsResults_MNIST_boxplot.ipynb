```python
parameters = df.columns[0:11]
parameters
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df['score_bin'] = pd.cut(
            df["adj_rand"], bins=np.linspace(0.0, 1.0, 6)
        )
```

```python
ncol=3
nrow=5
fig, axs = plt.subplots(ncol,nrow, figsize=(15,10))
fig.suptitle('Adjusted Rand')
fig.tight_layout(pad=2.0)

for i, var in enumerate(parameters):
    # print(var)
    # print(i//nrow, i%nrow)
    ax = axs[i//nrow, i%nrow]
    sns.boxplot(df, x=var, y='adj_rand', ax=ax)
    ax.set(title=var, xlabel='', ylabel='')
```

```python
ncol=3
nrow=5
fig, axs = plt.subplots(ncol,nrow, figsize=(15,10))
fig.suptitle('Proportion of noise')
fig.tight_layout(pad=2.0)

for i, var in enumerate(parameters):
    # print(var)
    # print(i//nrow, i%nrow)
    ax = axs[i//nrow, i%nrow]
    sns.boxplot(df, x=var, y='proportion_of_noise', ax=ax)
    ax.set(title=var, xlabel='', ylabel='')
```
